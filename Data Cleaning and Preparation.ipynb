### 2. Check and handle missing values

```python
df.isnull().sum()

# Only 'CouponCode' column contains missing values, and they are 309.
```

```python
df.fillna('No Coupon',inplace = True)

df

# If someone didn't used the coupon that's why it may be empty
# We handled the missing values in 'CouponCode' column by filling it as 'No coupon'. 
```

### 3. Check and handle duplicates

```python
df.duplicated().sum()

# There is no duplicate value so we do not need to handle duplicate values
```

#### - Check duplicates in 'OrderID'

```python
df['OrderID'].duplicated().sum()

# NO duplicate was found in 'OrderID'.
```

### 4. Verify Date Format

```python
df['Date'].dtype

# 'Date' column was already formatted
```

### 5.Numeric Precision

```python
df['UnitPrice'] = df['UnitPrice'].round(2)
df['TotalPrice'] = df['TotalPrice'].round(2)
```

```python
df[['UnitPrice', 'TotalPrice']].tail()
# Verifying it to see if the data formatting worked.
```

### 6. Validate numeric columns.

```python
df[['Quantity','UnitPrice','TotalPrice','ItemsInCart']].describe()


# No negative or unrealistic value was found.
```

```python
df = df[df['Quantity'] > 0]

# It makes the 'Quantity' more than 0.
```

### 7. Clean text Columns

```python
df.select_dtypes(include = 'object').columns

# This gives us all the text columns.
```

```python
for col in df.select_dtypes(include = 'object').columns:
    df[col] = df[col].str.strip()

# It removes the unwanted spaces in the text columns.
```

### 8. Final Verifications

```python
assert df.isnull().sum().sum() == 0 #Checking null values
assert df.duplicated().sum() == 0 #Checks for duplicated values
assert df['OrderID'].duplicated().sum() == 0 #Unique OrderID check
assert (df['TotalPrice'].round(2) == (df['Quantity'] * df['UnitPrice']).round(2)).all() #Total price consistency
assert (df['Quantity'] > 0).all() #Valid quantity check
assert (df['UnitPrice'] > 0).all() #Valid price check
```
